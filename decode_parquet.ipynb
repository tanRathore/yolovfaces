decode and extract from the parquet file 

```python
import pandas as pd
import os
import base64
from PIL import Image
import io
file_path = "/Users/tanishqsingh/Desktop/yolovfaces/edge_case_images_set2.parquet"  
df = pd.read_parquet(file_path)
print(df.columns)
image_column = "base64"  #column containing the encoded images
output_dir = "/Users/tanishqsingh/Desktop/yolovfaces/images"  
os.makedirs(output_dir, exist_ok=True)
for idx, row in df.iterrows():
    try:
        # Extracting base64 string
        image_data = row[image_column]

        # Decoding base64 string to bytes
        image_bytes = base64.b64decode(image_data)

        # Converting the bytes  of the pqrqet to an image
        image = Image.open(io.BytesIO(image_bytes))

        # Saved our images to the same folder as the on where i had saved the images for face recognition so dont have to complicate stuff 
        image_path = os.path.join(output_dir, f"image_{idx}.png")
        image.save(image_path)
        print(f"Saved: {image_path}")
    except Exception as e:
        print(f"Error processing row {idx}: {e}")
print("Extraction complete!")
```
